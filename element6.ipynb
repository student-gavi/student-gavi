```python
# prompt: 간단한 막대그래프를 만들어줘

import matplotlib.pyplot as plt

x = ['A', 'B', 'C', 'D', 'E']
y = [5, 2, 7, 8, 2]

plt.bar(x, y)
plt.show()
```

```python
# prompt: 위에 있는 막대그래프 색깔을 주황색으로 바꿔줘

import matplotlib.pyplot as plt
plt.bar(x, y, color='orange')
plt.show()
```

```python
# prompt: 원그래프를 하나 그려줘

import matplotlib.pyplot as plt
plt.pie(y, labels=x)
plt.show()
```
